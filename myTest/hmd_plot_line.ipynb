```python
import pandas as pd
```

```python
data = pd.read_excel('./group(1).xlsx')
```

```python
data.head()
```

```python
gg = data.groupby(['group1', 'year'])['tobinq'].mean()
```

```python
g1 = gg[0]
```

```python
g1, type(g1), g1.shape
```

```python
g2 = gg[1]
```

```python
g2, type(g2), g2.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 9), dpi=150)
g1.plot(ax=ax, legend=True, marker='o', markersize=10, linewidth=3, label='Control', fontsize=24)
g2.plot(ax=ax, legend=True, marker='o', markersize=10, linewidth=3, label='Treatment', fontsize=24)
ax.set_xlabel('年份', fontsize=26)
ax.set_ylabel('Tobinq', fontsize=26)
ax.legend(fontsize=24)
```

- 下面第二行引号里面填写路径和名字，./代表的是当前目录

```python
fig.tight_layout()
fig.savefig('./result.jpg')
```
